```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder 
import pickle
import numpy as np
from tensorflow.keras.models import load_model
```

```python
## load model 
model = load_model('model.h5')

with open('ohe_gen.pickle','rb') as file:
    ohe_gen = pickle.load(file)

with open('ohe_geo.pickle','rb') as file:
    ohe_geo = pickle.load(file)

with open('scaler.pickle','rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_encoded = ohe_geo.transform([[input_data['Geography']]])
geo_encoded = pd.DataFrame(geo_encoded,columns=ohe_geo.get_feature_names_out(['Geography']))
geo_encoded
```

```python
gen_encoded = ohe_gen.transform([[input_data['Gender']]])
gen_encoded = pd.DataFrame(gen_encoded,columns=ohe_gen.get_feature_names_out(['Gender']))
gen_encoded
```

```python
input_df = pd.DataFrame([input_data])
```

```python
input_df.drop(columns=['Gender','Geography'],inplace=True)
```

```python
input_df = pd.concat([input_df,gen_encoded,geo_encoded],axis=1,ignore_index=False)
input_df
```

```python
input_scaled = scaler.transform(input_df)
```

```python
prediciton=model.predict(input_scaled)
```

```python
print(prediciton)
```

```python
if prediciton >0.5:
    print("Likely to churn")
else:
    print("Will not churn")
```
